```python
from langchain_ollama import ChatOllama

llm = ChatOllama(
    model="qwen3.5",
    temperature=0
)
```

```python
messages = [
    (
        "system",
        "You are a helpful assistant that translates English to Frensh. Translate the user sentence."
    ),
    ("human","I love Programming.")
]
```

```python
ai_msg = llm.invoke(messages)
```
